## The .apply( ) Method with Columns Values

```python
def add_million(number):
    return str(number) + " Million"
```

```python
columns = ['Box_Office','Budget','Bond_Actor_Salary']
for i in columns:
    df[i] = df[i].apply(add_million)   
```

```python
df.head()
```

```python
def add(string):
    return string + " Awesome"
```

```python
col = ['Actor', 'Director']
for i in col:
    df[i] = df[i].apply(add)
```

```python
df.head()
```

## The .apply( ) Method with Rows Values

```python
def good_movies(row):
    actor = row[1]
    budget = row[4]
    if actor == 'Sean Connery':
        return 'The best movie'
    elif actor == 'David Niven' and budget>518.5:
        return 'Enjoyable'
    else:
        return 'Dont watch the move at all'
df.apply(good_movies,axis='columns')
```
